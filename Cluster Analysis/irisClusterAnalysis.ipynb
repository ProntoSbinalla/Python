# Iris Dataset Cluster Analysis

## Importing Libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn import preprocessing
```

## Importing Data

```python
data = pd.read_csv("../Datasets/iris_dataset.csv")

data.head()
```

```python
plt.scatter(data['sepal_length'], data['sepal_width'])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
```

```python
x = data.iloc[:, 0:2]

kmeans = KMeans(2)
kmeans.fit(x)

data['Cluster'] = kmeans.fit_predict(x)

data.head()
```

```python
plt.scatter(data['sepal_length'], data['sepal_width'], c = data['Cluster'], cmap = 'rainbow')
```

```python
x_scaled = preprocessing.scale(x)
```

```python
wcss = []

for i in range(1, 11):
    kmeans_scaled = KMeans(i)
    kmeans_scaled.fit(x_scaled)
    wcss.append(kmeans_scaled.inertia_)

wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python
kmeans_scaled = KMeans(5)

kmeans_scaled.fit(x_scaled)

data['Cluster'] = kmeans_scaled.fit_predict(x_scaled)

data.head()
```

```python
plt.scatter(data['sepal_length'], data['sepal_width'], c = data['Cluster'], cmap = 'rainbow')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```
